```python
%%sql
select location, population, max(total_cases) as "Highest No. of Cases", 
max((total_cases/population))*100 as "Percent of Population Infected"
from covid_deaths where continent is not null group by location, population order by 4 desc
```

```python
%%sql 
select location, max(cast(total_deaths as int)) as "Total Death Count" from covid_deaths where continent is not null 
group by location order by 2 desc 
```

```python
%%sql
select continent, max(cast(total_deaths as int)) as "Total Death Count" from covid_deaths where continent is not null 
group by continent order by 2 desc
```

```python
%%sql 
select date, sum(new_cases) as "Total Cases", sum(cast(new_deaths as int)) as "Total Deaths", 
(sum(cast(new_deaths as int))/sum(new_cases))*100 as "Death Percentage"
from covid_deaths where continent is not null group by date order by 1,2
```

```python
%%sql 
select sum(new_cases) as "Total Cases", sum(cast(new_deaths as int)) as "Total Deaths", 
(sum(cast(new_deaths as int))/sum(new_cases))*100 as "Death Percentage"
from covid_deaths where continent is not null order by 1,2
```

```python
%%sql
select * from covid_deaths death join covid_vaccinations vac on death.location=vac.location and death.date=vac.date limit 10
```

```python
%%sql
select death.continent, death.location, death.date, death.population, vac.new_vaccinations from covid_deaths death
join covid_vaccinations vac on death.location=vac.location and death.date=vac.date where death.continent is not null 
order by 2,3 limit 5000
```

```python
%%sql
select death.continent, death.location, death.date, death.population, vac.new_vaccinations, 
sum(cast(vac.new_vaccinations as int)) over (Partition by death.location order by death.location, death.date) 
as "Rolling People Vaccinated" from covid_deaths death join covid_vaccinations vac on death.location=vac.location 
and death.date=vac.date where death.continent is not null order by 2,3 limit 500
```

```python
%%sql
with PopvsVac (Continent, Location, Date, Population, New_Vaccinations, Rolling_People_Vaccinated) as (
select death.continent, death.location, death.date, death.population, vac.new_vaccinations, 
sum(cast(vac.new_vaccinations as int)) over (Partition by death.location order by death.location, death.date) 
as "Rolling People Vaccinated" from covid_deaths death join covid_vaccinations vac on death.location=vac.location 
and death.date=vac.date where death.continent is not null)
select *, (Rolling_People_Vaccinated/Population)*100 from PopvsVac limit 500
```
